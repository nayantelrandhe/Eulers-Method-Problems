## Lotka Volterra equations
The Lotka Volterra (predator prey) equations are a set of coupled differential equations.
$$
\\
$$
Solve the given set for the given constants $r = 2, m = 3, a = 4, b = 5$. At $t = 0$ start of with equal values of predator and prey $P = H = 10$.
Also obtain the phase diagram for the given equations.


Lotka-Volterra Equations
$$
\frac{d P}{d t}=-P m+b H P \\
\frac{d H}{d t}=H r-a H P \\
$$

$$
\\
\\
\begin{cases}P=P(t) \quad \text {  Number of Predators } \\
H=H(t) \quad \text { Number of Prey }\end{cases}
$$

$$
\\
\\
\\
\begin{cases}
r>0 & \text { Birth Rate of Prey } \\
m>0 & \text { Death Rate of Predators } \\
a>0 & \text { Death Rate of Prey/Predator } \\
b>0 & \text { Birth Rate of Predators/Prey }\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%%time
#constants
r=2
m=3
a=4
b=5

t=0         #initial value of time
s=0.0001    #step size for t
E=22        #final value of t to plot the solution upto

#initial values p and h
p=10
h=12



def Pdot(p,h):
    return (-m*p)+(b*h*p)              #dP/dt


def Hdot(p,h):
    return (h*r)-(a*h*p)               #dH/dt

P=[p]
H=[h]
T=[t]

while t<=E:
    pi=p
    p=p+(s*Pdot(p,h))
    P=P+[p]
    h=h+(s*Hdot(pi,h))
    H=H+[h]
    t=t+s
    T=T+[t]
```

```python
# P vs t and H vs t plot

plt.figure(figsize=(20,5), dpi=100)
plt.title("Lotka Volterra")
plt.xlabel("Time t")
plt.ylabel("Count")
plt.plot(T,P,'r',label='Predator',linewidth=0.8)
plt.plot(T,H,'b',label='Prey',linewidth=0.8)
plt.grid()
plt.legend(ncol=2)
plt.show()

print('\nP(t)=',p,', H(t)=',h,', t=',t)
```

```python
#Phase Diagram
plt.figure(figsize=(10,5), dpi=100)
plt.title("Phase space")
plt.xlabel("Prey H")
plt.ylabel("Predator P")
plt.plot(H,P,linewidth=0.8)
```
