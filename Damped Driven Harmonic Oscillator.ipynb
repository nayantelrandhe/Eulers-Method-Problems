## Damped Driven Harmonic Oscillator
##### The equation of a damped driven harmonic oscillator is given below. Given the following initial conditions, at initial time, position x = 0.2 and velocity = 0.8,   b = 0.2 and A = 0.1. Resonant frequency = 1.2 and driving frequency 1. Draw the position vs time graph. Find out if the total energy is increasing or decreasing
$$
\ddot{x}=-a x-b \dot{x}+A \sin \left(\omega_{0} t\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

#initial values
x=0.2
xdot=0.8
b=0.2
A=0.1
a=(1.2)**2     #a=(Resonant frequency)**2

t=0
h=0.01

DriF=1

E=100




def Xddot(x,xdot,t):
    return (-a*x)-(b*xdot)+(A*np.sin(DriF*t))

X=[x]
T=[t]
Xdot=[xdot]

while t<E:
    x=x+(h*xdot)
    t=t+h
    xdot=xdot+(Xddot(x,xdot,t)*h)
    X=X+[x]
    T=T+[t]
    Xdot=Xdot+[xdot]

plt.figure(figsize=(10,5), dpi=100)
plt.title("Damped Driven Harmonic Oscillator")
plt.xlabel("t (Time)")
plt.ylabel("x (Position)")
plt.plot(T,X)
plt.grid()
```

```python
X=np.array(X)
T=np.array(T)
Xdot=np.array(Xdot)

plt.figure(figsize=(10,5), dpi=100)
plt.title("Damped Driven Harmonic Oscillator  E/m vs t")
plt.ylabel("E_m (Energy parameter)")
plt.xlabel("t (Time)")
m=1                                 #mass
E_m=(Xdot**2/2)+(a*m*X**2/2)        #E_m=(v^2/2)+((k/m)^2*x^2/2)  i.e. (KE/m)+(PE/m)
plt.plot(T,E_m)
```

If the mass is an unit mass, E vs t curve will be of the above form

The Energy decreases initially, it seem to increase then decrease and periodically vary while it seem to oscillate after t=40, which coincides with steady state curve in the 1st graph.
